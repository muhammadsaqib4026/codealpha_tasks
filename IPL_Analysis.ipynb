```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:/Code_Alpha_Projects/EDA_Project/archive (1)/IPL_matches.csv")
```

```python
df = df.drop(columns = ['umpire3'], inplace = True)
```

```python
df['city'] = df['city'].fillna('unknown')
```

```python
df['winner'] = df['city'].fillna('No Result')
```

```python
df['umpire1'] = df['umpire1'].fillna('unknown')
```

```python
df['umpire2'] = df['umpire2'].fillna('unknown')
```

```python
df.isnull().sum()
```

```python
df['date'] = pd.to_datetime(df['date'], dayfirst=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['team1'] = df['team1'].str.strip()
```

```python
df['team2'] = df['team2'].str.strip()
```

```python
# VISUALIZATIONS
```

```python
df['winner'].value_counts().head().plot(kind = 'bar')
plt.title('Top Winning Teams')
plt.show()
```

```python
df['season'].value_counts().plot(kind = 'line')
plt.title('Matches Per Year')
plt.show()
```

```python
# IPL History shows that matches per season, almost changing every year , but the highest number of matches 
# played in IPL was in 2013
```

```python
df['toss_decision'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Toss Decision Distribution")
plt.show()
```

```python
# Graph shows that most of the teams choose to Bowl, after winning the toss
```

```python
plt.scatter(df['win_by_runs'], df['win_by_wickets'])
plt.title("Runs vs Wickets Wins")
plt.xlabel("Win by Runs")
plt.ylabel("Win by Wickets")
plt.show()
```

```python
# This visualization shows the fundamental rule that team can win either by wickets or by runs, but not by both
```
